## **Data Cleaning**

Here I will seperate the Crash Date and the Crash Time into two seperate columns. The crash Date will also be seperated by month, day, and year.

```python
df["Crash_Month"] = pd.to_datetime(df["Crash Date/Time"]).dt.month
df["Crash_Day"] = pd.to_datetime(df["Crash Date/Time"]).dt.day
df["Crash_Year"] = pd.to_datetime(df["Crash Date/Time"]).dt.year
df["Crash_Time"] = pd.to_datetime(df["Crash Date/Time"]).dt.time

df.head()
```

My research question is what time of day are crashes most likely to occur in Montgomery Country Maryland. This data set gives me information about crash reports from Montgomery County itself which includes information about the time each crash occured.

```python

Count_time = pd.Series(df["Crash_Time"]).value_counts()

summary_stats_time = pd.Series(df["Crash_Time"]).describe()

print("Summary Statistics")
print(summary_stats_time)
print("Count of Crashes by Time of Day")
print(Count_time)
```

```python
df["Crash_Time_in_hours"] = df["Crash_Time"].apply(lambda x: x.hour + x.minute/60 + x.second/3600)
sns.histplot(df["Crash_Time_in_hours"], bins=24)
plt.xlabel("Time of Day (Hours)")
plt.ylabel("Number of Crashes")
plt.title("Distribution of Crashes by Time of Day")
plt.show()
```

The histogram shows the distribution of crashes by the time of day.  Looking at the distribution it seems to be skewed left with most of the crashes occuring later in the day between 15:00 to 20:00 hours (3 pm to 8 pm) reaching to over 600 crashes.
