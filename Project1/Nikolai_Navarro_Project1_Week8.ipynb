## **Data Cleaning**

Here I will seperate the Crash Date and the Crash Time into two seperate columns. The crash Date will also be seperated by month, day, and year.

```python
df["Crash_Month"] = pd.to_datetime(df["Crash Date/Time"]).dt.month
df["Crash_Day"] = pd.to_datetime(df["Crash Date/Time"]).dt.day
df["Crash_Year"] = pd.to_datetime(df["Crash Date/Time"]).dt.year
df["Crash_Time"] = pd.to_datetime(df["Crash Date/Time"]).dt.time

df.head()
```

**My research question is what time of day are crashes most likely to occur in Montgomery Country Maryland?** This data set gives me information about crash reports from Montgomery County itself which includes information about the time each crash occured.

```python

Count_time = df["Crash_Time"].value_counts()

summary_stats_time = df["Crash_Time"].describe()

print("Summary Statistics")
print(summary_stats_time)
print("Count of Crashes by Time of Day")
print(Count_time)
```

```python
df["Crash_Time_in_hours"] = df["Crash_Time"].apply(lambda d: d.hour + d.minute/60 + d.second/3600)
sns.histplot(df["Crash_Time_in_hours"], bins=24)
plt.xlabel("Time of Day (Hours)")
plt.ylabel("Number of Crashes")
plt.title("Distribution of Crashes by Time of Day")
plt.show()
```

```python
Crash_time_mean = df["Crash_Time_in_hours"].mean()
Crash_time_median = df["Crash_Time_in_hours"].median()
Crash_time_mode = df["Crash_Time_in_hours"].mode()

print("Mean:", Crash_time_mean)
print("Median:", Crash_time_median)
print("Mode:", Crash_time_mode)
```

The histogram shows the distribution of crashes by the time of day. Looking at the distribution it seems to be skewed left with most of the crashes occurring later in the day between 16:00 to 20:00 hours (4 pm to 8 pm) reaching to over 600 crashes at its peak. This data would make sense as that is when most cars are on the road due to rush hour and many people driving home leading to more opportunities of crashes happening. In the earlier hours between 1 am to 6 am there seems to be little amounts of crashes happening which would make sense as most people would be asleep at those times.

**Another question is what month do most crashes occur?**

```python
month_counts = df["Crash_Month"].value_counts().sort_index()

#import calandar to change numbers into month names (help from Gemeni)
import calendar
month_names = [calendar.month_name[m] for m in month_counts.index]


summary_stats_month = df["Crash_Month"].describe()

print("Summary Statistics")
print(summary_stats_month)
print("Count of Crashes by Time of Day")
print(month_counts)
```

```python
sns.barplot(x=month_names, y=month_counts.values)
plt.title("Number of Crashes by Month")
plt.xlabel("Month")
plt.ylabel("Number of Crashes")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
Crash_month_mean = df["Crash_Month"].mean()
Crash_month_median = df["Crash_Month"].median()
Crash_month_mode = df["Crash_Month"].mode()

print("Mean:", Crash_month_mean)
print("Median:", Crash_month_median)
print("Mode:", Crash_month_mode)
```

Looking at the bar plot a most of the crashes occur in the later half of the year between August and December. The month with the most number of crashes is in October with over 777 crashes being reported. The first half of the year has a lower amount of crashes occurring with only January and June having a similar amount of reports as the later months. The spring season seems to have the lowest amount of crashes reported while the fall/winter season seems to have the most.

### d. Nonparametric Inference

- Use a random sample (~10%)
- Explain your method

Discuss:

- Results of the inference
- Interpretation

```python
month_count = df["Crash_Month"].value_counts()
sample_size = int(0.1 * len(df))
num_samples = 10000
sample_means = []

np.random.seed(42)
for _ in range(num_samples):
  sample = np.random.choice(month_count, size=sample_size, replace=True)
  sample_means.append(np.mean(sample))

plt.axvline(x=np.mean(month_count), color = "orange", ymin = 0, ymax = 1, linestyle = '--')

sns.histplot(sample_means, kde=True, bins = 30)
plt.title("Simulated Distribution of Mean Monthly Crashes")
plt.xlabel("Simulated Mean of Monthly Crashes")
plt.ylabel("Frequency")
plt.show()

print(f'Bootstrapped Mean: {np.mean(sample):.4}')
print(f'Bootstrapped Std: {np.std(sample):.4}')
```

The simulation shows a normal distribution. The distribution of the simulation is reaches its peak around 595 crashes which is around the actual mean of monthly crashes at 595.6. Because the data gathers around the actual mean this information shows us that during each month it is likely to expect around 595 crashes be reported.
