##Part A -- Translation (Conceptual)

1. Predict the output
2. Translate it into Python
3. Compute the result in Python

```r
x <- c(1, 3, 5, 7)
y <- ifelse(x > 4, x^2, x + 1)
mean(y)
```
Clearly show your Python code and final result.

* You first are creating a vector called x with numbers inside.
* Next you make a vector called y that looks at the numbers in vector x. If the number is greater than 4 than it is squared otherwise it gets 1 added to it.
* The last thing is getting the mean of the y vector.



````python
import numpy as np
x = np.array([1, 3, 5, 7])
y = np.where(x > 4, x**2, x + 1)
print(y)
np.mean(y)
````
